```python
import pandas as pd
from scipy.io.arff import loadarff

import arff
```

```python
data = loadarff('protein.arff')
df = pd.DataFrame(data[0])
df['class'] = df['class'].apply(lambda x: x.decode())
df
```

```python
df = (df.groupby('class', as_index=False)
        .apply(lambda x: x.sample(n=100, random_state=2220))
        .reset_index(drop=True))
df
```

```python
attributes = [(j, 'NUMERIC') if df[j].dtypes in ['int64', 'float64'] else (j, df[j].unique().astype(str).tolist()) for j in df]

arff_dic = {
  'attributes': attributes,
  'data': df.values,
  'relation': 'protein',
  'description': '',
}

with open('protein_bal.arff', 'w', encoding='utf8') as f:
  arff.dump(arff_dic, f)
```
