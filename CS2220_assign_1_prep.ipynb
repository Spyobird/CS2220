```python
import pandas as pd

import arff

from Bio import SeqIO
from Bio.SeqUtils.ProtParam import ProteinAnalysis
```

```python
alpha_records = []
beta_records = []
for record in SeqIO.parse('assignment1-alpha_seq-1.fasta', 'fasta'):
  prot = ProteinAnalysis(record.seq)
  d = prot.get_amino_acids_percent()
  d['class'] = 'alpha'
  alpha_records.append(d)
for record in SeqIO.parse('assignment1-beta_seq-1.fasta', 'fasta'):
  prot = ProteinAnalysis(record.seq)
  d = prot.get_amino_acids_percent()
  d['class'] = 'beta'
  beta_records.append(d)
alpha_df = pd.DataFrame(alpha_records)
beta_df = pd.DataFrame(beta_records)
df = pd.concat([alpha_df, beta_df], ignore_index=True)
df
```

```python
attributes = [(j, 'NUMERIC') if df[j].dtypes in ['int64', 'float64'] else (j, df[j].unique().astype(str).tolist()) for j in df]

arff_dic = {
  'attributes': attributes,
  'data': df.values,
  'relation': 'protein',
  'description': '',
}

with open('protein.arff', 'w', encoding='utf8') as f:
  arff.dump(arff_dic, f)
```
